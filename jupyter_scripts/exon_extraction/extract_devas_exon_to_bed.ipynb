```python
import csv
from collections import defaultdict
```

```python
# 初始化一个字典，键为组织名称，值为对应的行记录列表
bed_dict = defaultdict(list)
```

```python
a = list()
```

```python
with open('Bombyx_mori_as_output.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
       # tissue = row['tissue']
        # 提取需要的字段
        chr_val = row['chr']
        start = row['exonStart_0base']
        end = row['exonEnd']
        gene = row['GeneID']
        # 添加到对应组织的列表中
      #  bed_dict[tissue].append([chr_val, start, end, gene])
        a.append([chr_val, start, end, gene])
```

```python
with open(filename, 'w') as bedfile:
    for rec in a:
        bedfile.write("\t".join(rec) + "\n")
```
