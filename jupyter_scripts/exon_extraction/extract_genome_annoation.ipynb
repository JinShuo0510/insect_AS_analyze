```python
# 存储每个物种的转录本 ID，字典键为 species（例如 Bombyx_mori），值为 set 类型的 ID 集合
species_to_transcript_ids = {}
```

```python
# 读取正交组文件（假设文件为 tab 分隔）
with open(orthogroups_file, 'r') as infile:
    reader = csv.DictReader(infile, delimiter='\t')
    # 遍历每一行
    for row in reader:
        # 遍历每个字段（第一列为 Orthogroup，其余为各物种）
        for key, value in row.items():
            if key == "Orthogroup":
                continue  # 跳过第一列
            # 如果对应单元格为空，则跳过
            if not value.strip():
                continue

            # 物种名：将列名中的 ".protein" 去掉
            species = key.replace(".protein", "")
            if species not in species_to_transcript_ids:
                species_to_transcript_ids[species] = set()

            # 如果单元格中有多个转录本 ID，以逗号分隔（并去掉多余空格）
            ids = [tid.strip() for tid in value.split(",")]
            for tid in ids:
                species_to_transcript_ids[species].add(tid)
```

```python
# 用于从 GTF 属性中提取 transcript_id 的正则表达式
transcript_id_re = re.compile(r'transcript_id "([^"]+)"')
```

```python
## 遍历每个物种，处理对应的 GTF 文件
for species, tid_set in species_to_transcript_ids.items():
    # 构造 GTF 文件路径：假设在 /data/home/jinshuo/transcript/Genome/ 下，文件名为 {species}.gtf
    gtf_file = f"/data/home/jinshuo/transcript/Genome/{species}.gtf"
    if not os.path.exists(gtf_file):
        print(f"警告：文件 {gtf_file} 不存在，跳过 {species}！")
        continue

    # 输出文件（你可以根据需要修改输出路径或文件名）
    output_file = f"{species}.exon.gtf"
    with open(gtf_file, 'r') as gf, open(output_file, 'w') as out_f:
        for line in gf:
            # 跳过 GTF 文件的注释行
            if line.startswith("#"):
                continue
            fields = line.strip().split("\t")
            # 确保至少有 9 列
            if len(fields) < 9:
                continue
            feature = fields[2]
            # 只处理 exon 行
            if feature != "exon":
                continue
            attributes = fields[8]
            # 提取 transcript_id
            m = transcript_id_re.search(attributes)
            if not m:
                continue
            transcript_id = m.group(1)
            # 如果当前行的 transcript_id 存在于正交组中记录的该物种 ID 集合中，则输出此行
            if transcript_id in tid_set:
                out_f.write(line)
    print(f"物种 {species} 的 exon 行已提取到 {output_file}")
```
