```python
import csv
from collections import defaultdict
```

```python
# 构建Run到分组的映射
run_group_mapping = {}
with open('2024_04_01_DATA_withGroupinfo_with_treatment_and_description_sort_final_xifen.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        group = f"{row['Age']}__{row['Age_xifen']}__{row['Tissue.1']}"
        run_group_mapping[row['Run']] = group
```

```python
print(run_group_mapping)
```

```python
# 生成分组数据
with open('Bombyx_mori_rpkm_values.txt') as f:
    # 读取样本头
    sample_ids = next(f).strip().split()
    
    # 构建分组索引映射
    group_indices = defaultdict(list)
    for idx, sample in enumerate(sample_ids):
        if group := run_group_mapping.get(sample):
            group_indices[group].append(idx)
    
    # 输出表头
    print("GeneID\t" + "\t".join(group_indices.keys()))
    
    # 处理基因行
    for line in f:
        parts = line.strip().split()
        gene_id = parts[0]
        values = [float(x) for x in parts[1:]]
        
        # 构建输出行
        output = [gene_id]
        for group, indices in group_indices.items():
            group_values = [f"{values[i]:.4f}" for i in indices]
            output.append(",".join(group_values))
        
        print("\t".join(output))
```
